```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def NominalRes(image,size):
  maxLn = 0
  for y in range(image.shape[0]):
    ln = 0
    for x in range(image.shape[1]):
      if image[y,x] == 1:
        ln+=1
      else:
        if ln > maxLn:
          maxLn = ln
        ln = 0
  return maxLn/size

for i in range(1,7):
  file = open(f'figure{i}.txt')
  size = float(file.readline())
  file.readline()
  image = []
  for line in file:
    image.append([int(ch) for ch in line.split()])
  image = np.array(image)
  plt.imshow(image)
  plt.show()

  print(NominalRes(image,size))

  file.close()
```

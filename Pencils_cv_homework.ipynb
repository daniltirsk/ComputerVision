```python
import matplotlib.pyplot as plt
from skimage import filters
import numpy as np
import skimage
from skimage.measure import label, regionprops
from skimage.filters import try_all_threshold, threshold_local, threshold_isodata, threshold_mean, threshold_yen,threshold_otsu, median,gaussian
from skimage.filters import threshold_sauvola
```

```python
images = [plt.imread(f'img ({i}).jpg') for i in range(1,13)]
```

```python
binary_images = [np.mean(image,2) for image in images]
```

```python
for image in images:
  plt.subplot(121)
  plt.imshow(image)
  plt.subplot(122)
  #plt.imshow(filters.threshold_otsu(np.mean(image, 2)))
  plt.imshow(np.mean(image, 2))
  plt.show()
  print(np.mean(image,2).shape)
```

```python
img = np.mean(images[1],2)

fig, ax = try_all_threshold(img, figsize=(20, 20), verbose=False)
plt.show()
```

```python
thresholded_images = [image<threshold_otsu(image) for image in binary_images]
```

```python
for image in thresholded_images:
  plt.imshow(image)
  plt.show()
```

```python
pencil_count_total = 0
for image in thresholded_images:
  plt.imshow(image)
  plt.show()
  pencil_count_image = 0
  for region in regionprops(label(image)):
    if region.perimeter > 1500:
      if 30 > region.major_axis_length/region.minor_axis_length > 15:
        pencil_count_image+=1
  print(pencil_count_image)
  pencil_count_total+=pencil_count_image
```
